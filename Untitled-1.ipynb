```python
import requests

url = 'https://dl.fbaipublicfiles.com/segment_anything_2/072824/sam2_hiera_tiny.pt'
file_name = 'sam2_hiera_tiny.pt'

# Send a GET request to download the file
response = requests.get(url)

# Save the content of the response to a file
with open(file_name, 'wb') as file:
    file.write(response.content)

print(f"{file_name} downloaded successfully.")
```

```python
from sam2.build_sam import build_sam2
from sam2.sam2_image_predictor import SAM2ImagePredictor
from sam2.utils.misc import variant_to_config_mapping
from sam2.utils.visualization import show_masks

model = build_sam2(
    variant_to_config_mapping["tiny"],
    "/Users/wtdafg/Documents/GitHub/datathon-24/sam2_hiera_tiny.pt",
)
image_predictor = SAM2ImagePredictor(model)
image_predictor.set_image(image)
```

```python
input_point = np.array([[300, 600]])
input_label = np.array([1])
```

```python
masks, scores, logits = image_predictor.predict(
    point_coords=input_point,
    point_labels=input_label,
    box=None,
    multimask_output=True,
)
```

```python
output_mask = show_masks(image, masks, scores)
```

```python
sorted_ind = np.argsort(scores)[::-1]
print(f"Top Score: {scores[sorted_ind[0]]}")
```

```python
import torch
from segment_anything import SamPredictor, sam_model_registry
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# Check available model keys
print("Available SAM models:", sam_model_registry.keys())

# Load the SAM model using an available key
sam_checkpoint = "/Users/wtdafg/Documents/GitHub/datathon-24/sam2_hiera_tiny.pt"
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Use the correct model key
sam = sam_model_registry["default"](checkpoint=sam_checkpoint)  # Replace with the correct key
sam.to(device)

# Load the image
image_path = '/Users/wtdafg/Desktop/Screenshot 2024-10-12 at 17.49.30.png'
image = Image.open(image_path)
image_np = np.array(image)

# Prepare the SAM predictor
predictor = SamPredictor(sam)
predictor.set_image(image_np)

# You can define input points and labels if needed (interactive segmentation)
# For automatic segmentation, use default settings
masks, _, _ = predictor.predict(box=None, points=None)

# Visualize the results
plt.figure(figsize=(10, 10))
plt.imshow(image_np)
plt.imshow(masks[0], alpha=0.5)  # Display the first segmentation mask with transparency
plt.axis('off')
plt.show()
```
